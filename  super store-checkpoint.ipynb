```python
 types_of_customers=df['Segment'].unique()
```

```python
number_of_customer=df['Segment'].value_counts().reset_index()
print(number_of_customer)
```

<h4>pie chart</h4>

```python
plt.pie(number_of_customer['Segment'], labels=number_of_customer['index'] ,autopct='%1.1f%%')
plt.title('Customers')
plt.show()
```

```python
sales_per_category=df.groupby('Segment')['Sales'].sum().reset_index()
print(sales_per_category)
```

<h4>BAR GRAPH</h4>

```python
plt.bar(sales_per_category['Segment'],sales_per_category['Sales'])
plt.title('Sales per Category')
plt.xlabel('Segment')
plt.ylabel('Sales')
plt.show()
```
